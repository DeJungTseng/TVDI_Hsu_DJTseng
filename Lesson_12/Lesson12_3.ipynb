```python
from dotenv import load_dotenv
import os
import requests
load_dotenv()
def get_aqi_str(aqi_key:str)->str:
    url=f"https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key={aqi_key}&limit=1000&sort=ImportDate%20desc&format=JSON"
    try:
        r=requests.get(url)
        r.raise_for_status()
    except Exception as e:
        print(f"發生錯{e}")
    return r.text

# call the get_aqi_str function with arg (aqi_key)
# the aqi_key gets from os.environ['AQI_KEY']
aqi_data=get_aqi_str(os.environ['AQI_KEY'])
```

```python
# make every data as a class in AQI JSON file
# field_validator is a decorator to check whether 
from pydantic import BaseModel, Field, field_validator,field_serializer
from datetime import datetime

class Record(BaseModel):
    日期: datetime = Field(alias='datacreationdate')
    站點:str=Field(alias="sitename")
    縣市:str=Field(alias="county")
    aqi:float | None 
    狀態:str=Field(alias="status")
    pm25:float | None =Field(alias="pm2.5")
    緯度:float=Field(alias="latitude")
    經度:float=Field(alias="longitude")


    # ===Validate the input===
    # check the Null value input data by decorator
    # if there's no mode="before", exec the "pm25" and "aqi" attribute in the Class first
    @field_validator("pm25","aqi",mode='before')
    # apply a decorator @classmethod with 2 args (cls,v) 
    # and if the validator v ("pm25","aqi"column) is Null, fill the column with None
    @classmethod
    def pm25_for_empty_string(cls,v:str)->float:
        if v=="":
            return None
        else:
            return float(v)
    
    # ====Serialize the output===
    # modify the lat and lon into 4 digit
    @field_serializer("緯度","經度")
    def four_digits(self, v:float)->float:
        return round(v,ndigits=4)
    
    @field_serializer('日期')
    def datetime_to_string(self,date:datetime) -> str:
        return date.strftime('%Y:%m:%d %H:%M:00')

class AQI(BaseModel):
   records:list[Record]

aqi=AQI.model_validate_json(aqi_data)
```

```python
# change into python structure: Serialize

aqi.model_dump()['records']
for aqi_item in aqi_data:
    print(aqi_item)
```
